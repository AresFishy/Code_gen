```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from jupyterthemes import jtplot
jtplot.style('oceans16')
%matplotlib inline
```

```python
stim = 3 # stimulus
stim_std = 0.1 # stimulus standard deviation
prop = abs(np.random.randn())  # initial random guess
timelength = 2000
dt = 0.1
eps = 1e-4
```

```python
props = np.zeros(timelength)
targets = np.zeros(timelength)
targets[0] = stim
target0 = stim + stim_std * np.random.randn()
```

```python
for ind in range(1,timelength):
#     prop = prop + dt * (target0-prop) / (1 + target0 - np.mean(targets[:ind]))   #  / (1 + np.var(targets-props))
    clip_ind = np.clip(ind-10,0,ind-10)
    prop = prop + dt * (target0-prop) * (1 + target0 - np.mean(targets[clip_ind:ind]) + eps)
    if ind < timelength / 2:
        target0 = stim + stim_std*np.random.randn()
    else:
        target0 = stim - 0.4 + 4*stim_std*np.random.randn()
    props[ind] = prop
    targets[ind] = target0
    
# Incorporate "surprisal" into learning rule
```

```python
sns.distplot(props)
sns.distplot(targets)
```
